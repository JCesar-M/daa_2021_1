```
# 1. Suma recursiva de una lista de numeros enteros. 

import java.util.ArrayList;

public class SumaRecursiva_Lista {
 
	public static void main(String[] args) {
		
	
     ArrayList<Integer> listaNumeros = new ArrayList();
     
     listaNumeros.add(1);
     listaNumeros.add(3);
     listaNumeros.add(5);
     listaNumeros.add(6);
     listaNumeros.add(12);
     listaNumeros.add(13);
     listaNumeros.add(2);
     listaNumeros.add(9);
     
     System.out.println("\"Lista numeros\"");
     
     for (int i = 0; i < listaNumeros.size(); i++) {
		System.out.println(listaNumeros.get(i));
     }
  
     System.out.println("---------------------");
     System.out.println("La suma es: "+sumaRecursiva(listaNumeros)); 
     System.out.println("---------------------");
     
     System.out.println("\nVerificando si la lista esta vacia\n");
     
     if(listaNumeros.isEmpty()) {
    	 System.out.println("----->>>La lista esta vacia<<<------");
     }else {
    	 System.out.println("---->>>Hay elemento en la lista<<<<-------");
     }
  
	 }
	
	 static int suma = 0;
   
	 public static Integer sumaRecursiva(ArrayList<Integer> lista) {
		
		  int tamanio = lista.size(); // se guarda el tamaño de la lista
		
	    if(tamanio == 0){    // si la lista es vacia restorna la suma. 
	     	return suma; 
		  }
	    else{    // si la lista no esta vacia empieza a sumar.
	      suma += lista.get(tamanio-1);  // suma el ultimo numero. 
			  lista.remove(tamanio-1);   // elimina ese numero que sumamos.
			  sumaRecursiva(lista);   // se vuleve a llamar la funcion. 
		  }
		return suma;
	 }
	 
}
```

```
#2. Hacer un contador regresivo con recursión.

public class ContadorRegresivo {
	
	public static void main(String[] args) {
		
		System.out.println("Contador regresivo: \n");
		contadorRegresivo(20); 
		
	}

	public static void contadorRegresivo(int numero){
		if(numero == 0) { // Si el numero es igual a cero termina la recursividad. 
		   System.out.println(numero); // Se imprime el cero. 
		}else {
			System.out.println(numero); // se imprime el numero que recibe el parametro de la funcion.
			contadorRegresivo(numero-1); // se hace la recursividad. 
		}
	}
}
```
